```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data
```

# Numerical-Numerical Data

```python
# sns.scatterplot(data['Age'], data['Survived'])
sns.scatterplot(x=data['Embarked'], y=data['Fare'])
```

# Numerical-Categorical Data

```python
sns.barplot(x=data['Embarked'], y=data['Age'])
```

```python
sns.boxplot(x=data['Sex'] , y=data['Fare'])
```

```python
sns.distplot(data[data['Survived'] == 0]['Age'], hist=False)
sns.distplot(data[data['Survived'] == 1]['Age'], hist=False)
```

```python
sns.kdeplot(data=data[data['Survived'] == 0], x='Age', fill=True)
sns.kdeplot(data=data[data['Survived'] == 1], x='Age', fill=True)
```

# Categorical-Categorical Data

```python
pd.crosstab(data['Pclass'], data['Survived'])
```

```python
sns.heatmap(pd.crosstab(data['Pclass'], data['Survived']))
```

```python
sns.clustermap(pd.crosstab(data['Embarked'], data['Survived']))
```
